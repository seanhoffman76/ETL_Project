```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import the rotten_tomatoes_movies.csv to a DataFrame, and then output the fields and datatypes
# so the data can be easily mapped in an ERD.

movies_file = "rotten_tomatoes_movies.csv"
rot_movies_df = pd.read_csv(movies_file)
rot_movies_df.dtypes
```

```python
review_file = "rotten_tomatoes_reviews.csv"
rot_review_df = pd.read_csv(review_file)
rot_review_df.dtypes
```

```python
rot_movies_df.head(10)
```

```python
# Drop unnecessary fields from the movies table

new_rot_movies_df = rot_movies_df[['rotten_tomatoes_link',
                                   'movie_title',
                                   'critics_consensus',
                                   'rating',
                                   'in_theaters_date',
                                   'on_streaming_date',
                                   'studio_name',
                                   'tomatometer_rating',
                                   'tomatometer_count',
                                   'audience_rating',
                                   'audience_count']].copy()

new_rot_movies_df
```

```python
# Remove movies released after 2015-01-01 to match the data from the Oscars data set

new_rot_movies_df = new_rot_movies_df[new_rot_movies_df.in_theaters_date <= '2015-01-01']
new_rot_movies_df
```

```python
# Check for any duplicates in the new_rot_movies_df and remove them if found
new_rot_movies_df = new_rot_movies_df.drop_duplicates()
new_rot_movies_df
```

```python
# Remove movies have no tomatometer_rating value or no audience_rating value

new_rot_movies_df = new_rot_movies_df[new_rot_movies_df.tomatometer_rating >= 0]
new_rot_movies_df = new_rot_movies_df[new_rot_movies_df.audience_rating >= 0]
new_rot_movies_df
```

```python
new_rot_movies_df.to_csv('rotten_tomatoes.csv')
```
