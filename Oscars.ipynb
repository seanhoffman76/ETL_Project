```python
import pandas as pd
```

```python
#import csv with relevent utf-8
csv = "../Oscars-demographics-DFE.csv"
oscars_df = pd.read_csv(csv, encoding = 'latin1' )
oscars_df.head()
```

```python
#create a new dataframe that shows only the needed information
oscars_df1 = oscars_df[['person', 'award','movie','year_of_award']]
oscars_df1.head()
```

```python
#create new data frames categorized my award
director_df = oscars_df1[oscars_df1.award == 'Best Director']
actor_df = oscars_df1[oscars_df1.award == 'Best Actor']
actress_df = oscars_df1[oscars_df1.award == 'Best Actress']
supporting_actor_df = oscars_df1[oscars_df1.award == 'Best Supporting Actor']
supporting_actress_df = oscars_df1[oscars_df1.award == 'Best Supporting Actress']
```

```python
director_df = director_df.drop_duplicates()
actor_df = director_df.drop_duplicates()
actress_df = director_df.drop_duplicates()
supporting_actor_df = director_df.drop_duplicates()
supporting_actress_df = director_df.drop_duplicates()
```

```python
#export to csv's
oscars_df1.to_csv('clean_oscars.csv',header=True,index=True, encoding= 'latin1')
director_df.to_csv('director_oscars.csv',header=True,index=True, encoding= 'latin1')
actor_df.to_csv('actor_oscars.csv',header=True,index=True, encoding= 'latin1')
actress_df.to_csv('actress_oscars.csv',header=True,index=True, encoding= 'latin1')
supporting_actor_df.to_csv('supporting_actor_oscars.csv',header=True,index=True, encoding= 'latin1')
supporting_actress_df.to_csv('supporting_actress_oscars.csv',header=True,index=True, encoding= 'latin1')
```
