```python
import pandas as pd
```

```python
imdb_df = pd.read_csv("movie_data.csv")
```

```python
# create df and drop rows with na values
imdb_df = imdb_df.dropna()
imdb_df
```

```python
# drop duplicate rows in the movie_title column
imdb_df = imdb_df.drop_duplicates(subset="movie_title", keep="first")
imdb_df
```

```python
# pare down original DF to eliminate unnecessary columns
imdb_df2 = imdb_df[['director_name','movie_title','title_year','duration','genres','actor_1_name','actor_2_name','actor_3_name','movie_imdb_link','num_voted_users','imdb_score']]
imdb_df2
```

```python
# change the title_year column to an integer
imdb_df2['title_year']=imdb_df2['title_year'].astype(int).copy()
imdb_df2
```

```python
# change the duration column to an integer
imdb_df2['duration'] = imdb_df2['duration'].astype(int).copy()
imdb_df2
```

```python
# change the duration column to an integer
imdb_df2['num_voted_users'] = imdb_df2['num_voted_users'].astype(int).copy()
imdb_df2
```

```python
#exclude movies made before 1926 and after 2015
imdb_df3 = imdb_df2[(imdb_df2.title_year < 2015) & (imdb_df2.title_year >= 1970)] 
imdb_df3
```

```python
# use min/max line to confirm year filter worked
imdb_df3.title_year.max()
```

```python
# rename columns in DF
cleaned_imdb = imdb_df3.rename(columns={
    "director_name":"director",
    "movie_title":"title",
    "title_year":"year_released",
    "duration":"duration_min",
    "genres":"genres",
    "actor_1_name":"actor_1",
    "actor_2_name":"actor_2",
    "actor_3_name":"actor_3",
    "movie_imdb_link":"imdb_url",
    "num_voted_users":"review_count",
    "imdb_score":"imdb_score"
})
cleaned_imdb
```

```python
# check data types of DF to ensure compatibility with SQL table
cleaned_imdb.dtypes
```

```python
# export new DF into csv
cleaned_imdb.to_csv('clean_imdb.csv',header=True,index=False)
```
